```python
import pandas as pd
```

```python
data = pd.read_csv("modified_file.csv")
```

```python
data
```

```python
data['CUSINE'].unique()
```

```python
# Filter the rows where the CUSINE is 'Konkan'
Konkan_recipes = data[data['CUSINE'] == 'Konkan']

# Print the Konkan recipes
print(Konkan_recipes)
```

```python
# Replace 'Konkan' with 'West Indian' in the 'CUSINE' column
data['CUSINE'] = data['CUSINE'].replace('Konkan', 'West Indian')

# Save the modified dataframe to a new CSV file
output_file_path = 'modified_file.csv'  # Specify the output file name
data.to_csv(output_file_path, index=False)

print(f"File saved as {output_file_path}")
```
